Database Initialization and Data Loading

```python
import pandas as pd

file = "drugsdata.csv"

df = pd.read_csv(file)

print("Successfully loaded the Pharmacogenomics Database:")

print(df.head())
```

```python
df.loc[df['drug_name'].str.contains('Clopidogrel', case=False), 'gene'] = 'CYP2C19'

df.loc[df['drug_name'].str.contains('Aspirin', case=False), 'gene'] = 'PTGS1'

df.to_csv('drugsdata.csv', index=False)

print("Data correction complete. The file 'drugsdata.csv' has been updated successfully!")

print("Updated Database Preview:")
print(df)
```

Clinical Decision Support System

```python
search_drug = input("Enter the drug name to check PGx recommendations:")

result = df[df["drug_name"].str.contains(search_drug, case=False, na=False)]

if not result.empty:
    print("Genetic Recommendation Found.")
    print(f"Drug:{result["drug_name"].values[0]}")
    print(f"Associated Gene:{result["gene"].values[0]}")
    print(f"Medical Advice:{result["recommendation"].values[0]}")

else:
    print("No Data Found for '{search_drug}'.")
```

The Interaction Loop

```python
while True:
    search_drug = input("Enter drug name (or 'exit' to quit): ").strip()


    if search_drug.lower() == 'exit':
        print("Closing the system. Goodbye!")
        break


    result = df[df["drug_name"].str.contains(search_drug, case=False, na=False)]

    if not result.empty:    print("Genetic Recommendation Found.")
    print(f"Drug:{result["drug_name"].values[0]}")
    print(f"Associated Gene:{result["gene"].values[0]}")
    print(f"Medical Advice:{result["recommendation"].values[0]}")

else:
    print("No Data Found for '{search_drug}'.")
```
